```python
# prepare for sklearn
X = patsy.dmatrix("""~ C(embarked, Treatment(reference='S')) + C(sex, Treatment(reference='male')) 
              + C(p_class, Treatment(reference='3')) + sib_spouse + age""", data=titanic_scaled)
y = titanic.survived.values
```

```python
# fit model
log_reg = linear_model.LogisticRegression()
log_reg.fit(X, y)
```

```python
import math
```

```python
# examine coefficients
df = pd.DataFrame({'*features' : X.design_info.column_names, 
                   'model_coef': log_reg.coef_[0,:],
                  'odds_ratio': math.e**log_reg.coef_[0,:]})
df
```

```python
# predict class labels
log_reg_pred = log_reg.predict(X)
```

```python
# predict probabilities
log_reg_proba = log_reg.predict_proba(X)
```

```python
# evaluate model
print 'Model accuracy: ', log_reg.score(X, y)
```

The model with all the default values has an accuracy around 80%, which is quite a bit better than the baseline of 50%, or 61% if guessing always "not survived", but not great.

```python
# split and test generalization
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33, random_state = 109)
```

```python
# evaluate train model on test
log_reg.fit(X_train, y_train)
print 'Training model score on Test subset: ', log_reg.score(X_test, y_test)
```

The model fitted on the training dataset has a 79% accuracy, which is close to the accuracy of the model on itself, suggesting good generalization.

```python
# cross validate test set
scores = sklearn.model_selection.cross_val_score(log_reg, X, y, cv=5)
print 'cross-val scores: ', scores
print 'mean cross-val: ', scores.mean(), '+/-', scores.std()
```

Cross-validated accuracy is closer to 79% on average, and remains fairly consistent, suggesting good generalizability.

```python
y_pred = sklearn.model_selection.cross_val_predict(log_reg, X, y, cv=3)
print 'accuracy:', sklearn.metrics.accuracy_score(y, y_pred)
print 'precision:', sklearn.metrics.precision_score(y, y_pred)
print 'recall:', sklearn.metrics.recall_score(y, y_pred)
```

```python
# classification report
print sklearn.metrics.classification_report(y, y_pred)
```

#### What do the classification metrics tell us?

The classification metrics tell us:
- this model correctly predicts 85% of the casualties and 69% of the survivors. 
- 80% of the people this model predicted would be casualties were in fact casualties, and 76% of people who the model predicted would survive did in fact survive

It also tells us the harmonic balance of these metrics, which is 78% overall.

```python
# confusion matrix
actual_vs_cross_val_pred = pd.DataFrame([y_pred, y]).transpose()
actual_vs_cross_val_pred.columns = ['cross_val_pred', 'actual']
pd.crosstab(actual_vs_cross_val_pred.cross_val_pred, actual_vs_cross_val_pred.actual)
```

#### What does the Confusion Matrix tell us? 

The confusion matrix tells us that:
- The model correctly identified 469 passengers as dying
- The model correctly identified 230 passengers as surviving
- The model incorrectly identified 80 passengers as surviving when they really died
- The model incorrectly identified 110 passengers as dying when they really survived.

```python
# plot and explain the ROC curve
y_score = model_selection.cross_val_predict(log_reg, X, y, method='predict_proba')[:,1]
fpr, tpr, thresholds = sklearn.metrics.roc_curve(y, y_score)
print 'AUC:', sklearn.metrics.auc(fpr, tpr)

plt.plot(fpr, tpr, label='AUC: {:.2}'.format(sklearn.metrics.auc(fpr, tpr)))
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC Curve for Cross-Validated Model')
plt.plot([0, 1], [0, 1], 'k--')
plt.legend(loc='lower right')
```

The ROC curve is another way to measure model performance that takes into account how it performs at different thresholds. The Area Under the Curve (AUC) is 0.84, which is very good - a worthless model has a score of 0.5 and a perfect model has a score of 1.0, and this is much closer to 1 than .5.

## Part 5: Gridsearch
Use GridSearchCV with logistic regression to search for optimal parameters.

```python
# parameter dict
logreg_parameters = {
    'penalty' : ['l1','l2'],
    'C' : [.00001, .0001, .001, .01, .1, 1, 5, 10, 100, 1000, 10000]              }
best_log_reg = model_selection.GridSearchCV(log_reg, logreg_parameters)
best_log_reg.fit(X, y)
```

```python
# print best parameters and score; is it better than original model?
print 'best parameters: ', best_log_reg.best_params_
print 'best score: ', best_log_reg.best_score_
```

```python
# compute difference between L1 (Lasso) and L2 (Ridge) penalties on model coefficients
df = pd.DataFrame({'*features' : X.design_info.column_names, 
                   'L2_model_coefficients': log_reg.coef_[0,:],
                  'L2_odds_ratio': math.e**log_reg.coef_[0,:],
                  'L1_model_coefficients': best_log_reg.best_estimator_.fit(X, y).coef_[0,:],
                  'L1_odds_ratio': math.e**best_log_reg.best_estimator_.fit(X, y).coef_[0,:],
                  'l2_odds/l1_odds': math.e**log_reg.coef_[0,:]/math.e**best_log_reg.best_estimator_.fit(X, y).coef_[0,:]})
df
```

As you can see, the L1 regularization kicked out the predictors related to the number of parents/children passengers had on board and related to those who embarked from Queensland. Generally, the odds ratios are larger or about the same in L2 compared to L1, except the "base" odds (in this case, the odds for an average-aged. 3rd-class male who departed from Southampton who paid the average fair and had no siblings, parents, spouses, or children on board) were significantly larger for L1 than L2.

#### BONUS: Thresholds
You decide that you want to minimize false positives. Use the predicted probabilities from the model to set your threshold for labeling the positive class to need at least 90% confidence. How and why does this affect your confusion matrix?

```python
y_prob = model_selection.cross_val_predict(log_reg, X, y, cv=3, method='predict_proba')
y_prob
```

```python
thresh_90 = pd.DataFrame(y_prob, columns=['prob_died', 'prob_survived'])
thresh_90['actually_survived'] = y
thresh_90['predicted_survived_90'] = thresh_90.prob_survived.apply(lambda x: 1 if x>=.9 else 0)
thresh_90.head()
```

```python
#new confusion matrix and classification table
print metrics.classification_report(thresh_90.actually_survived, thresh_90.predicted_survived_90)
pd.crosstab(thresh_90.predicted_survived_90, thresh_90.actually_survived)
```

With a threshold set to 90, this model:
- correctly identified 546 people as dying who actually died
- correctly identified 55 people as surviving who actually survived
- incorrectly identified 285 people as dying who actually survived
- incorrectly identified 3 people as surviving who actually died

That last one is critical here: the new threshold did create a lot of incorrect predictions (specifically, there were many people the model predicted would die but who actually survived), but the goal of this is threshold change was to minimize the number of people who we thought would survive who actually died and that's just what happened. In fact, less than 1% of the casualties went unpredicted. Unfortunately, this is because we predicted most people (93%) would die, and had lots of false positives as a result. If we wanted to ensure we intervene in 99% of the people who would otherwise die while not intervening on basically everyone, we'd need a better model.

```python
tree_params = {
    'criterion': ['gini', 'entropy'],
    'splitter':['best', 'random'],
    'max_depth':np.arange(1,20,2),
    'min_samples_split':(2,100,20),
    'random_state':[1,5],
    'min_samples_leaf': [1,3,5,9,11],
    'max_features':[2,4,6,8],
}
tree_select = model_selection.GridSearchCV(tree.DecisionTreeClassifier(), param_grid=tree_params, scoring='roc_auc')
tree_select.fit(X, y)
```
